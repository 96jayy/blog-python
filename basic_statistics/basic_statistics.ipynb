```python
for i in range(data.shape[1]):
    if data.iloc[:,i].dtype=='O':
        print(data.columns[i],'unique values :', data.iloc[:,i].unique())
        print(data.columns[i], 'unique counts :', data.iloc[:,i].nunique())
        print('\n')
```

```python
def unique_var(data):
    for i in range(data.shape[1]):
        if data.iloc[:,i].dtype=='O':
            print(data.columns[i],'unique values :', data.iloc[:,i].unique())
            print(data.columns[i], 'unique counts :', data.iloc[:,i].nunique())
            print('\n')

unique_var(data)
```

날짜로 된 데이터는 날짜 변환 후 데이터 확인하기

```python
import warnings

# UserWarning만 무시
warnings.filterwarnings('ignore', category=UserWarning)
for col in data.columns:
    if data[col].dtype=='O':
        try:
            # 날짜 변환 시도
            converted = pd.to_datetime(data[col], errors='raise', infer_datetime_format=True)

            # 변환 성공하면, 해당 컬럼을 날짜 타입으로 업데이트
            data[col] = converted
            print(f"✅ '{col}' 컬럼은 날짜 데이터로 변환되었습니다.")
        except (ValueError, TypeError):
            print(f"❌ '{col}' 컬럼은 날짜 데이터가 아닙니다.")
data.info()
```

```python
def date_var(data):
    for col in data.columns:
        if data[col].dtype=='O':
            try:
                # 날짜 변환 시도
                converted = pd.to_datetime(data[col], errors='raise', infer_datetime_format=True)

                # 변환 성공하면, 해당 컬럼을 날짜 타입으로 업데이트
                data[col] = converted
                print(f"✅ '{col}' 컬럼은 날짜 데이터로 변환되었습니다.")
            except (ValueError, TypeError):
                print(f"❌ '{col}' 컬럼은 날짜 데이터가 아닙니다.")
    print(data.info())
date_var(data)
```

```python
def detect_categorical_columns(df, threshold=0.05):
    cat_cols = []
    for col in df.columns:
        if df[col].dtype == 'object':
            unique_ratio = df[col].nunique() / len(df[col])
            if unique_ratio < threshold:
                cat_cols.append(col)
                print(f"✅ '{col}' → 범주형 (고유값 비율: {unique_ratio:.2%})")
            else:
                print(f"ℹ️ '{col}' → 일반 문자열 (고유값 비율: {unique_ratio:.2%})")
    return cat_cols
```

```python
detect_categorical_columns(data)
```
